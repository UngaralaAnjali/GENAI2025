```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.randint(0,100,10),columns=['Sales'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,7))
plt.plot(df['Sales'],marker='o',color='#dc33ff',markerfacecolor='w',
         markeredgecolor='black',markeredgewidth=2,linewidth=10,alpha=0.9)
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Sales vs Index")
```
